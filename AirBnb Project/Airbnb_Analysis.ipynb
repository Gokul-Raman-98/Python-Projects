```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('datasets.csv', encoding_errors = 'ignore')
```

```python
# Initial Exploration
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
# Statistical Summary
data.describe()
```

```python
# Data Cleaning

data.isnull().sum()
```

```python
# Dropping missing values rows

data.dropna(inplace = True)

data.isnull().sum()
```

```python
data.shape
```

```python
# Dealing with duplicate rows

data.duplicated().sum()

data[data.duplicated()]
```

```python
# Deleting duplicate rows

data.drop_duplicates(inplace = True)
data.duplicated().sum()
```

```python
#changing data types

data.dtypes

data['id'] = data['id'].astype(object)
data.dtypes

data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

```python
# Univarite Analysis

# Identify the outliers in price field

sns.boxplot(data=data, x='price')
```

```python
# creating dataframe with the outliers removed

df = data[data['price'] < 1500]

sns.boxplot(data= df, x='price')
```

```python
# Find the Distribution of Price column

plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins = 100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
# Find the Distribution of availability_365 column

plt.figure(figsize=(6,3))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
# Find the average price for each neighbourhood
df.groupby(by ='neighbourhood_group')['price'].mean()
```

```python
# Feature Engineering
df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
# Find the average price per bed for each region
df.groupby(by ='neighbourhood_group')['price per bed'].mean()
```

```python
# Bi Variable Analysis

sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# number of reviews and price relation

plt.figure(figsize=(8,5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
# Geographical Distribution of AirBnb Listings

plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listings")
plt.show()
```

```python
# Heatmap - correlation of one variable with others for numerical column

corr = df[['latitude','longitude','price','minimum_nights', 'number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr

plt.figure(figsize=(8,6))
sns.heatmap(data=corr, annot=True)
```
